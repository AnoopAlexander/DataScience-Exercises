## Using Matplotlib to manipulate how data is visualized

# Create sample data using numpy array

```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
import matplotlib.pyplot as plt
%matplotlib inline
```

# create a simple plot using object oriented style in matplotlib

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title')
```

# Create a plot within the above graph 

```python
fig1 = plt.figure()
ax1=fig1.add_axes([0,0,1,1])
ax1.plot(x,y)
ax1.set_xlabel('x')
ax1.set_ylabel('y')

ax2=fig1.add_axes([0.2,0.6,0.4,0.4])
ax2.plot(x,y)
ax2.set_xlabel('x')
ax2.set_ylabel('y')
```

## creation of subplots

```python
fig, axes = plt.subplots(nrows=1, ncols=2,figsize=(12,2))

axes[0].plot(x,y,color="blue", lw=5)
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')

axes[1].plot(x,z,color="red", lw=3, ls='--')
axes[1].set_xlabel('x')
axes[1].set_ylabel('z')
```
